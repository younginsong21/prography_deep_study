```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn import svm

# 데이터 세팅
data = np.array([[1,2],
             [5,8],
             [1.5,1.8],
             [8,8],
             [1,0.6],
             [9,11]])
# xy 분리
x = data[:, 0]
y = data[:, 1]
plt.scatter(x, y)
plt.show()

# 각 좌표가 속한 label 세팅
label = [0, 1, 0, 1, 0, 1]

# SVM classifier 세팅
svm_clf = svm.SVC(kernel='linear', C=1.0)
svm_clf.fit(data, label)

# 새로운 좌표가 들어왔을 때 label 예측
# 아마 new1은 lable:0에, new2은 label:1로 예측될거임
new1 = [[0.58, 0.76]]
new2 = [[10.58, 10.76]]
print("new1:", svm_clf.predict(new1))
print("new2:", svm_clf.predict(new2))
```

```python
# 데이터와 svm classifier line을 함께 그래프에 나타내기

w = svm_clf.coef_[0]

svm_a = -w[0] / w[1]
svm_b = -svm_clf.intercept_[0] / w[1]

svm_x = np.linspace(0, 12)
svm_y = svm_a * svm_x + svm_b

grp = plt.plot(svm_x, svm_y, 'k-', label='svm line')
plt.scatter(x, y)
plt.legend()
plt.show()
```
